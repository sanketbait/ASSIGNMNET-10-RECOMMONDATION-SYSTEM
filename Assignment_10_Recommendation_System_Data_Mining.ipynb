```python
# IMPORT THE LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# load the dataset
books = pd.read_csv("/content/book.csv", encoding='Latin1')
books
```

```python
# As we observed there are extra index columns in our dataset hence removing 
books2=books.iloc[:,1:]
books2
```

```python
# Sort by User IDs
books2.sort_values(['User.ID'])
```

```python
# number of unique users in the dataset
len(books2['User.ID'].unique())
```

```python
# number of unique books in the dataset
len(books2['Book.Title'].unique())
```

```python
# number of unique Ratings in the dataset
len(books2['Book.Rating'].unique())
```

```python
#Displaying userid numbers based on ratings count
books2['Book.Rating'].value_counts()
```

```python
#displaying book title based on value counts 
books2['Book.Title'].value_counts()
```

```python
# converting long data into wide data using pivot table
books3 = books2.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books3
```

```python
# Replacing the index values by unique user Ids
books3.index=books2['User.ID'].unique()
books3
```

```python
#Impute those NAN with 0 values 
books3.fillna(0,inplace=True)
books3
```

```python
# Calculating Cosine Similarity between Users on array data
user_cosine=1-pairwise_distances(books3.values,metric='cosine')
user_cosine
```

```python
# Store the results in a dataframe format
user_cosine2_df=pd.DataFrame(user_cosine)
user_cosine2_df
```

```python
#Setting the user names and columns to user id 
user_cosine2_df.index=books2['User.ID'].unique()
user_cosine2_df.columns=books2['User.ID'].unique()
user_cosine2_df
```

```python
# Nullifying diagonal values 
np.fill_diagonal(user_cosine,0)
user_cosine2_df
```

```python
#checking most similar users 
user_cosine2_df.idxmax(axis=1)
```
